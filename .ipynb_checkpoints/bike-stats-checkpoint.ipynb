### Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
df_weather.head
```

```python
df_weather.Date = pd.to_datetime(df_weather.Date)
```

```python
df_weather['DAY'] = df_weather.Date.apply(dt.datetime.isoweekday)
df_weather['DAY_OF_WEEK'] = df_weather.Date.dt.weekday_name
df_weather['MONTH'] = df_weather.Date.dt.month
df_weather['YEAR'] = df_weather.Date.dt.year
```

```python
df_weather.head
```

```python
grouped = df_weather.Precipitation_In.groupby(df_weather.MONTH)
```

```python
series = grouped.count()
series
```

```python
df_weather['Precipitation_In'].median()
```
